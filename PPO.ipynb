```python
from environment.fogg_behavioural_model import Patient 
from stable_baselines3.common.env_checker import check_env
```

```python
env1 = Patient()
env_info1 = {'behaviour_threshold': 20, 'has_family':True}
env1.env_init(env_info1)
```

```python
check_env(env1)
```

```python
from stable_baselines3 import PPO


model = PPO("MlpPolicy", env1, verbose=1, tensorboard_log="./patient1_tensorboard/")
model.learn(total_timesteps=25000)
model.save("patient1")
```

```python
env1.observation_space
```

```python
obs = env1.reset()
x = []
y = []
for i in range(0,100):
    x.append(obs)
    action, _states = model.predict(obs)
    obs, rewards, dones, info = env1.step(action)
    y.append(rewards)
```
